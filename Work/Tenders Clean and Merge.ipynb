```python
import pandas as pd
import os

# 1. Define your folder path
folder_path = '/Users/matei/Documents/Classes/Economics/ECO 358R/Paper Proposal #1 (EU Aid v. Single Bids)/Data/Tenders'

cleaned_dataframes = []

# Sort files to process them in chronological order
files = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])

print(f"Scanning directory: {folder_path}\n")

# 2. Loop through every pre-cleaned CSV
for filename in files:
    filepath = os.path.join(folder_path, filename)
    print(f"Processing {filename}...")

    try:
        # Load the file. Since it only has ~6 variables, memory is no longer an issue.
        df = pd.read_csv(filepath, dtype={'TAL_LOCATION_NUTS': str})

        # --- APPLY FILTERS ---

        # Filter 1: Drop rows where NUMBER_OFFERS is missing
        df_clean = df.dropna(subset=['NUMBER_OFFERS']).copy()

        # Filter 2: Keep only NUTS codes with exactly 4 characters
        # We strip whitespace first to ensure codes like "ES61 " aren't accidentally dropped
        df_clean['TAL_LOCATION_NUTS'] = df_clean['TAL_LOCATION_NUTS'].astype(str).str.strip()
        df_clean = df_clean[df_clean['TAL_LOCATION_NUTS'].str.len() == 4]

        # Store the filtered dataframe
        cleaned_dataframes.append(df_clean)
        print(f"  -> Kept {len(df_clean)} valid contracts.")

    except Exception as e:
        print(f"  -> ERROR processing {filename}: {e}")

# 3. Combine and Save
if cleaned_dataframes:
    print("\nCombining all files...")
    master_df = pd.concat(cleaned_dataframes, ignore_index=True)

    output_filename = os.path.join(folder_path, "TED_Master_Cleaned_2007_2020.csv")
    master_df.to_csv(output_filename, index=False)

    print(f"\nSuccess! Master dataset saved as:\n'{output_filename}'")
    print(f"Total usable contracts for your model: {len(master_df)}")
else:
    print("\nFailure: No dataframes to combine.")
```
