```python
import pandas as pd

# 1. Load the math you already did in Excel
df_agri = pd.read_excel("Sectoral_Composition.xlsx", sheet_name="Agri_Share")

# 2. FORCE the first column's name to be 'Region'
df_agri.rename(columns={df_agri.columns[0]: 'Region'}, inplace=True)

# 3. Melt it in Python!
df_agri_clean = df_agri.melt(id_vars='Region', var_name='Year', value_name='Agriculture_Share')

# 4. Save to Excel
df_agri_clean.to_excel("Final_Agri_Controls.xlsx", index=False)
```
